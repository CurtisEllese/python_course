| В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего из 1 столбца. Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без get_dummies? |

```python
import random
```

```python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
df = pd.DataFrame({'whoAmI':lst})
df.head()
```

```python
pd.get_dummies(data=df, columns=['whoAmI'], drop_first=True)
```

```python
pd.get_dummies(df['whoAmI'])
```

```python
for index, value in df['whoAmI'].items():
  if value == 'robot':
    df.loc[index, 'whoAmI_robot'] = 1
    df.loc[index, 'whoAmI_human'] = 0
  if value == 'human':
    df.loc[index, 'whoAmI_human'] = 1
    df.loc[index, 'whoAmI_robot'] = 0

df['whoAmI_robot'] = df['whoAmI_robot'].astype(int)
df['whoAmI_human'] = df['whoAmI_human'].astype(int)
```

```python
# Очистим таблицу
df = df.loc[:, :'whoAmI']

# Можно не учитывать первый столбец, так как легко предсказывается на основе предыдущего

for index, value in df['whoAmI'].items():
  if value == 'robot':
    df.loc[index, 'whoAmI_robot'] = 1
  if value == 'human':
    df.loc[index, 'whoAmI_robot'] = 0

df['whoAmI_robot'] = df['whoAmI_robot'].astype(int)
```
